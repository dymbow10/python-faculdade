```python
import pandas as pd
df = pd.read_excel("houses_to_rent.xlsx")
```

```python
#usando o sort values
df.sort_values("total").head() 
#se eu passar o parametro inplace = True em qualqer comando eu estou alterando o dado em si
#ex: df.sort_values("total", inplace = True)
```

```python
#ordenando por mais de uma coluna
#a prioridade de ordenação vai da esquerda pra direita, ou seja, quando os valores da esquerda são iguais
#ele ordena pelo da direita
df.sort_values(["total", "area"]).head()
```

```python
#usando ascending e descending obs: descending é só ascending = False
df.sort_values("total", ascending = False).head()
```
